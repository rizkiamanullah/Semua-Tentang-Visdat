```python
import networkx as nx
import matplotlib.pyplot as plt
```

Inisiasi nodes kota-kota besar di Jawa Barat

```python
G = nx.Graph()
G.add_nodes_from(
    [
        'Bandung',
        'Banjar',
        'Bekasi',
        'Bogor',
        'Cimahi',
        'Cirebon',
        'Depok',
        'Sukabumi',
        'Tasikmalaya'
        ]
)
G.nodes()
nx.draw(G, with_labels=True)
plt.show()
```

Pembuatan edges antar kota besar di Jawa Barat berdasarkan jalan nasional dari gambar referensi dibantu dengan Google Maps

```python
G.add_edge('Bandung', 'Sukabumi')
G.add_edge('Bandung', 'Cirebon')
G.add_edge('Bandung', 'Bekasi')
G.add_edge('Bandung', 'Cimahi')
G.add_edge('Bandung', 'Tasikmalaya')
G.add_edge('Banjar', 'Tasikmalaya')
G.add_edge('Bekasi', 'Bogor')
G.add_edge('Bekasi', 'Depok')
G.add_edge('Bogor', 'Sukabumi')
G.add_edge('Depok', 'Bogor')
G.add_edge('Cimahi', 'Sukabumi')
G.edges()
nx.draw(G, with_labels=True)
plt.show()
```

Degree centrality

```python
nx.degree_centrality(G)
```

Betweenness centrality

```python
nx.betweenness_centrality(G)
```

Dari hasil Degree Centrality dan Betweenness Centrality berikut adalah 5 kota terpenting untuk menghubungkan kota-kota besar di Jawa Barat

1. Bandung
2. Bekasi
3. Bogor
4. Tasikmalaya
5. Sukabumi
